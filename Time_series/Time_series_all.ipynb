```python
precipitation_2005 = boulder_precip_2003_2013.loc['2005']
precipitation_2005.head()
```

```python
# Remove missing data values
precip_2005_clean = precipitation_2005.dropna()
```

```python
# Plot the data using pandas

precip_2005_clean.reset_index().plot(x="DATE",
                                     y="HPCP",
                                     title="Hourly Precipitation",
                                     kind="scatter")
plt.show()
```

There is another way to plot scatter


```python
# Plot the data using native matplotlib
f, ax = plt.subplots()
ax.scatter(x=precip_2005_clean.index.values,
           y=precip_2005_clean["HPCP"])
plt.show()
```

```python
precip_2005_daily = precip_2005_clean.resample("D").sum()
# Plot the data using native matplotlib
f, ax = plt.subplots()
ax.scatter(x=precip_2005_daily.index.values,
           y=precip_2005_daily["HPCP"])
plt.show()
```

```python
prec_2012 = boulder_precip_2003_2013.loc['2012']
prec_2013 = boulder_precip_2003_2013.loc['2013']
prec_2012.resample("D").max() #for summerizing day data  in 2012 or 2013
prec_2013.resample("D").max()
fig,(ax1,ax2) = plt.subplots(2,1, figsize=(12,8))
ax1.scatter(x=prec_2012.index.values,
           y=prec_2012["HPCP"],)

ax2.scatter(x=prec_2013.index.values,
           y=prec_2013["HPCP"],)
ax1.set_title("Hourly Precipitation in 2012")
ax2.set_title("Hourly Precipitation in 2013")
plt.subplots_adjust(hspace=0.5) #adjust space for vertical spacing between two figure
plt.show()
```

```python
prec_2012 = boulder_precip_2003_2013.loc['2012']
prec_2013 = boulder_precip_2003_2013.loc['2013']
prec_2012.resample("D").max() #for summerizing day data  in 2012 or 2013
prec_2013.resample("D").max()
fig,(ax1,ax2) = plt.subplots(2,1, figsize=(12,8))
ax1.scatter(x=prec_2012.index.values,
           y=prec_2012["HPCP"], color ='green')

ax2.scatter(x=prec_2013.index.values,
           y=prec_2013["HPCP"], color='green')
ax1.set_title("Hourly Precipitation in 2012", color='black')
ax2.set_title("Hourly Precipitation in 2013", color='black')
ax1.set_ylim(0,2)
ax2.set_ylim(0,2)
ax1.set_ylabel("Precipitation (in)")
ax2.set_ylabel("Precipitation (in)")
ax1.set_xlabel("Time(Year)")
ax2.set_xlabel("Time(Year)")
plt.subplots_adjust(hspace=0.5) #adjust space for vertical spacing between two figure
plt.show()
```

```python
from matplotlib.dates import DateFormatter

# Place your code to plot your data here
flood_data = boulder_precip_2003_2013['2013-09-01':'2013-11-01']

f, ax = plt.subplots(figsize=(10, 6))

ax.scatter(x=flood_data.index.values,
           y=flood_data["HPCP"])

# Define the date format
date_form = DateFormatter("%m-%d")
ax.set_xlabel("Date")
ax.set_ylabel("Precipitation (in)")
ax.xaxis.set_major_formatter(date_form)
ax.set(title="Precipitation Sept - Nov 2013")
plt.show()
```
