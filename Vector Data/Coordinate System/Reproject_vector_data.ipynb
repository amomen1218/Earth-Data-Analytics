```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import earthpy as et

# Setting plotting style for the notebook
sns.set_style("white")
sns.set(font_scale=1.5)

roads = gpd.read_file('/content/drive/MyDrive/Earth data Analytics/earth-analytics/data/spatial_vector_lidar/california/madera-county-roads')
aoi_shp = gpd.read_file('/content/drive/MyDrive/Earth data Analytics/earth-analytics/data/spatial_vector_lidar/california/neon-sjer-site/vector_data/SJER_crop.shp')
print(roads.crs)
print(aoi_shp.crs)
```

```python
aoi_reproject = aoi_shp.to_crs(roads.crs)
print(aoi_reproject.crs)
aoi_reproject.head()
```

```python
fig, ax= plt.subplots(figsize=(10,9))
roads.plot(cmap='grey', ax=ax, alpha=.5)
aoi_reproject.plot(color='red', ax=ax, markersize=10)
ax.set_title('Roads map with AOI')
```
